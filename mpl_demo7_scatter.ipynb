```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
plt.style.use('seaborn')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/'\
                   '2019-05-31-data.csv')
data.head(5)
```

```python
viev_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(viev_count, likes, c = ratio, cmap = 'summer' ,edgecolor = 'black', linewidth = 1, alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
```
